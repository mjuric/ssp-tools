```python
from lsst.daf.butler.formatters.parquet import pq, astropy_to_arrow
from tqdm import tqdm

# Prepare output directory
output_file = f"dia_source_visit_combined.parquet"

# Stream DiaSources directly to a single Parquet file
writer = None
nsrc = {}

for dataRef in tqdm(dataRefs):
    dataId = dataRef.dataId
    
    # Get the table from butler
    table = butler.get(dataRef)
    nsrc[dataId["visit"]] = len(table)
    
    # Convert to Arrow table
    arrow_table = astropy_to_arrow(table)
    
    # Initialize writer with schema from first batch
    if writer is None:
        writer = pq.ParquetWriter(
            output_file,
            arrow_table.schema,
            compression='zstd',
            use_dictionary=False
        )
    
    # Write this batch as a row group
    writer.write_table(arrow_table)

# Close the writer
if writer is not None:
    writer.close()

print(f"Total sources written: {sum(nsrc.values()):,}")
print(f"Output file: {output_file}")
```

```python
from lsst.daf.butler.formatters.parquet import pq, astropy_to_arrow
from tqdm import tqdm
import numpy as np

def write_tables_to_parquet(table_generator, output_file, compression='zstd', total=None):
    """
    Write tables from a generator to a single Parquet file.
    
    Parameters
    ----------
    table_generator : generator
        Generator that yields tables
    output_file : str
        Path to output Parquet file
    compression : str
        Compression algorithm (default: 'zstd')
    total : int, optional
        Total number of items for progress bar
    
    Returns
    -------
    int
        Total number of rows written
    """
    writer = None
    total_rows = 0
    
    for table in tqdm(table_generator, total=total):
        # Skip if no rows
        if len(table) == 0:
            continue
        
        # Convert to Arrow table
        arrow_table = astropy_to_arrow(table)
        
        # Initialize writer with schema from first batch
        if writer is None:
            writer = pq.ParquetWriter(
                output_file,
                arrow_table.schema,
                compression=compression,
                use_dictionary=False
            )
        
        # Write this batch as a row group
        writer.write_table(arrow_table)
        total_rows += len(table)
    
    # Close the writer
    if writer is not None:
        writer.close()
    
    return total_rows


def query_and_yield_tables(repo, collection, datasetType, filter_ids=None):
    """
    Query datasets from Butler and yield filtered tables.
    
    Parameters
    ----------
    repo : str
        Path to Butler repository
    collection : str
        Collection name
    datasetType : str
        Dataset type to query
    filter_ids : array-like, optional
        Array of IDs to filter by (filters on diaSourceId if provided)
    
    Returns
    -------
    generator
        Generator that yields filtered astropy tables
    int
        Total number of dataRefs
    """
    import lsst.daf.butler as dafButler
    
    # Initialize butler
    butler = dafButler.Butler(repo, collections=collection)
    
    # Query datasets
    dataRefs = list(butler.registry.queryDatasets(
        datasetType=datasetType,
        collections=collection
    ))
    
    def _generator():
        # Process each dataRef
        for dataRef in dataRefs:
            # Get the table from butler
            table = butler.get(dataRef)
            
            # Apply filter if provided
            if filter_ids is not None:
                mask = np.isin(table['diaSourceId'], filter_ids)
                filtered_table = table[mask]
            else:
                filtered_table = table
            
            yield filtered_table
    
    return _generator(), len(dataRefs)


# Example usage
obssubid_array = df['obssubid'].values.astype(np.int64)
print(f"Filtering for {len(obssubid_array):,} unique obssubid values")

output_file = "dia_source_visit_filtered.parquet"

# Create the generator and write to Parquet
table_gen, num_datasets = query_and_yield_tables(
    repo='/repo/main',
    collection='LSSTCam/runs/DRP/FL/w_2025_19/DM-50795',
    datasetType='dia_source_visit',
    filter_ids=obssubid_array
)

total_written = write_tables_to_parquet(table_gen, output_file, total=num_datasets)

print(f"Total sources written: {total_written:,}")
print(f"Output file: {output_file}")
```

```python
import pyarrow.parquet as pq
import pyarrow as pa
import numpy as np

def parquet_to_structured_array(parquet_file, batch_size=100000):
    """
    Memory-efficiently read a Parquet file into a structured numpy array.
    Pre-allocates the array to avoid memory doubling from concatenation.
    Works directly with Arrow, avoiding pandas.
    Strings are loaded into fixed-width character columns.
    
    Parameters
    ----------
    parquet_file : str
        Path to the Parquet file
    batch_size : int
        Number of rows to read per batch (default: 100000)
    
    Returns
    -------
    numpy.ndarray
        Structured numpy array with all data
    """
    # Open the Parquet file
    parquet_file_obj = pq.ParquetFile(parquet_file)
    
    # Get total number of rows
    total_rows = parquet_file_obj.metadata.num_rows
    
    # Get the schema to build the numpy dtype
    arrow_schema = parquet_file_obj.schema_arrow
    
    # First pass: determine max length for string columns
    string_max_lengths = {}
    for field in arrow_schema:
        if pa.types.is_string(field.type) or pa.types.is_large_string(field.type):
            string_max_lengths[field.name] = 0
    
    if string_max_lengths:
        print("Scanning for max string lengths...")
        for batch in parquet_file_obj.iter_batches(batch_size=batch_size):
            for col_name in string_max_lengths.keys():
                arrow_col = batch.column(col_name)
                np_col = arrow_col.to_numpy(zero_copy_only=False)
                # Filter out None values and get max length
                valid_strings = [s for s in np_col if s is not None]
                if valid_strings:
                    max_len = max(len(s) for s in valid_strings)
                    string_max_lengths[col_name] = max(string_max_lengths[col_name], max_len)
        print(f"String column max lengths: {string_max_lengths}")
    
    # Convert Arrow schema to numpy dtype
    dtype_list = []
    for field in arrow_schema:
        arrow_type = field.type
        # Map Arrow types to numpy types
        if pa.types.is_int64(arrow_type):
            np_type = np.int64
        elif pa.types.is_int32(arrow_type):
            np_type = np.int32
        elif pa.types.is_int16(arrow_type):
            np_type = np.int16
        elif pa.types.is_float64(arrow_type):
            np_type = np.float64
        elif pa.types.is_float32(arrow_type):
            np_type = np.float32
        elif pa.types.is_boolean(arrow_type):
            np_type = np.bool_
        elif pa.types.is_timestamp(arrow_type):
            # Map timestamp to datetime64 with matching precision
            unit = arrow_type.unit  # 's', 'ms', 'us', or 'ns'
            np_type = np.dtype(f'datetime64[{unit}]')
        elif pa.types.is_date32(arrow_type) or pa.types.is_date64(arrow_type):
            np_type = np.dtype('datetime64[D]')
        elif pa.types.is_string(arrow_type) or pa.types.is_large_string(arrow_type):
            # Use fixed-width string with max length found
            max_len = string_max_lengths.get(field.name, 1)
            np_type = f'U{max_len}'  # Unicode string with fixed width
        else:
            print(f"Unknown type: {field}, {arrow_type}")
            # Default to object for complex types
            np_type = object
        
        dtype_list.append((field.name, np_type))
    
    dtype = np.dtype(dtype_list)
    
    # Pre-allocate the full array
    result = np.empty(total_rows, dtype=dtype)
    
    # Second pass: read the file in batches and fill the array
    print("Loading data...")
    offset = 0
    for batch in parquet_file_obj.iter_batches(batch_size=batch_size):
        batch_size_actual = len(batch)
        
        # Convert each column from Arrow to numpy and assign to structured array
        for col_name in batch.schema.names:
            arrow_col = batch.column(col_name)
            # Convert Arrow array to numpy
            np_col = arrow_col.to_numpy(zero_copy_only=False)
            # Assign to the structured array slice
            result[col_name][offset:offset + batch_size_actual] = np_col
        
        offset += batch_size_actual
    
    return result

# Read the Parquet file into a structured array
dia_sources = parquet_to_structured_array('dia_source_visit_filtered.bak.parquet')

print(f"Loaded {len(dia_sources):,} rows")
print(f"Columns: {dia_sources.dtype.names}")
print(f"Memory usage: {dia_sources.nbytes / 1024**2:.2f} MB")
```
